El objetivo de la práctica es agrupar y arreglar los datos para que un algoritmo de aprendizaje pueda leerlos y entenderlos
y así cuando el algoritmo reciba un test podrá predecir su valor.
    
En primer lugar importaremos todo lo necesario y utilizaremos panda para leer el dataset:

```python
import os

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(os.path.join("..", "in", "fifa.csv"))
pd.set_option('display.max_columns', None)
df.head()
```

El proceso que vamos a llevar a cabo es el siguiente:
> - Eliminar las columnas que no aporten ningún tipo de información
> - Buscar los jugadores que no pueden pasar por el algoritmo debido a falta de datos
> - Corregir datos que pueden ser relevantes pero no están expresados de forma correcta para el algoritmo

Bajo estos puntos, lo primero que haremos será eliminar las dos primeras columnas ya que los identificadores no aportan nada.

```python
df = df.iloc[:, 2:]
df.head()
```

Acto seguido sacamos de la tabla los datos que son completamente irrelevantes tales como la foto, la bandera, el logo del club...
Asumimos que, además de las columnas que son imágenes o nombres (las cuales lógicamente no aportan ninguna información), las columnas que contienen datos sobre el contrato no tienen impacto en el algoritmo puesto que conocer las fechas de duración de un contrato no hará que un jugador que tenga altas estadísticas se vea afectado y, por ejemplo, un "rookie" puede ser muy bueno o muy malo independientemente de si acaba de entrar en un equipo o no.

```python
df.pop("Photo")
df.pop("Flag")
df.pop("Loaned From")
df.pop("Club Logo")
df.pop("Jersey Number")
df.pop("Name")
df.pop("Real Face")
df.pop("Contract Valid Until")
df.pop("Joined")
df.head()
```

El siguiente paso será hacer un descarte de un bajo porcentaje de los jugadores a los cuales no se les puede aplicar el algoritmo debido a falta de datos. 
Mediante la siguiente codificación buscaremos aquellos jugadores que no tienen club, posición o altura. El por qué de ello es simple, estos jugadores representan un porcentaje muy bajo de la muestra, además los que no tienen altura coinciden en su mayoría con jugadores de los cuales no tenemos muchos datos (tales como peso, posición...).


```python
df = df.dropna(subset = ['Club'])
df = df.dropna(subset = ['Position'])
df = df.dropna(subset = ['Height'])
df.head()
```

A continuación vamos a reordenar ciertos datos para que el algoritmo pueda entenderlos. En este caso sabemos que tener una columna de clubes, nacionalidades... No es muy relevante para el algoritmo y que a este no le sirve leer un string con el nombre de un club por ejemplo. Así que nuestro siguiente paso será convertir cada club, nacionalidad, tipo de cuerpo... En una columna distinta y añadiremos un 1 si el jugador pertenece a alguna de esas características o un 0 en su defecto.
Las columnas seleccionadas para este proceso son: club, nacionalidad, pie preferente, tipo de cuerpo, posición y Work Rate.

Durante el proceso he tenido que investigar ciertos aspectos de algunas de las columnas. Un ejemplo de esto es el work rate. El work rate es un valor que se le da al ataque/defensa del jugador según su estilo de juego. Que un jugador tenga un work rate alto/alto no quiere decir que sea mejor que uno que lo tenga medio/medio, simplemente cambia su posición en el campo según si es atacante o defensor. Bajo este criterio he dividido el work rate según como atacan y como defienden. Seguiremos el mismo criterio que con el resto y añadiremos un 1 si el jugador contiene esta característica o un 0 si no.

```python
clb = df.pop("Club")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)

nat = df.pop("Nationality")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(nat, prefix='nat').reset_index(drop=True)], axis=1, sort=False)

foot = df.pop("Preferred Foot")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(foot, prefix='foot').reset_index(drop=True)], axis=1, sort=False)

bot = df.pop("Body Type")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(bot, prefix='bot').reset_index(drop=True)], axis=1, sort=False)

pos = df.pop("Position")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)

wr = df.pop("Work Rate")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(wr, prefix='wr').reset_index(drop=True)], axis=1, sort=False)

df.head()
```

Finalmente vamos a normalizar las columnas que a pesar de que son relevantes para el algoritmo, no están expresadas de forma óptima. 
Empezaremos por las cifras monetarias en las cuales hemos de multiplicar según si son millones (M) o miles (K). Además hemos de eliminar el símbolo del euro para que el algoritmo entienda bien el número.

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    if(x != x): return 0
    
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
df.head()
```

Siguiendo con el proceso anterior normalizamos también los pesos y las alturas, ya que una de ellas contiene la unidad (lbs) y la otra el carácter " ' " el cual debemos pasar a decimal normal.

```python
def peso_a_float(x):
    """ 
    Convertimos el peso a un float
    
    """ 
    x = float(x.replace("lbs", ""))
    return x
```

```python
def altura_a_float(x):
    """
    Convertimos la altura en un float
    
    """   
    x = x.split("'")
    
    if(len(x[1]) == 1): x = float(x[0]) + float(x[1])/10
    else:
        y = float(x[1])/100
        x = float(x[0])+y
        
    return x
```

```python
df["Weight"] = df["Weight"].apply(peso_a_float)
df["Height"] = df["Height"].apply(altura_a_float)
df.head()
```

Una vez más vamos a arreglar las últimas columnas mal expresadas. En este caso las columnas de estadísticas para los jugadores las cuales tienen la forma de x+y. Esta suma representa las estadísticas del jugador más un modificador. He supuesto que el modificador depende de la predilección del jugador a ciertas estadísticas, así que sumar el valor no tendrá ningún impacto en el algoritmo y así conseguimos que pueda leerlo.

En este apartado hemos podido observar en clase que tenemos un problema con los porteros, ya que no tienen estadísticas registradas en muchos aspectos. La solución tomada ha sido hacer una media de las estadísticas y colocárselo a los porteros de forma que el algoritmo no los desheche automáticamente, ya que el no tener estadísticas de tiro (por ejemplo) no influye en que un portero pueda ser muy bueno y valer mucho más dinero que un buen tirador.

> - En primer lugar sumamos las estadísticas excluyendo las estadísticas NaN, las cuales dejaremos tal cual.
> - En segundo lugar sustituiremos los valores NaN por una media de las estadísticas mediante la función "replace".

```python
def add_stats(x):
    """
    Suma las estadísticas y sustituye las columnas NaN por la media
    
    """

    if(x != x): return np.nan
        
    
    stats = x.split('+')
    
    
    return (float(stats[0]) + float(stats[1]))
```

```python
stats = ["LS","ST","RS","LW","LF","CF","RF","RW","LAM","CAM","RAM","LM","LCM","CM","RCM","RM","LWB","LDM","CDM","RDM","RWB","LB","LCB","CB","RCB","RB"]
for st in stats: df[st] = df[st].apply(add_stats)
df.head()
```

```python
stats = ["LS","ST","RS","LW","LF","CF","RF","RW","LAM","CAM","RAM","LM","LCM","CM","RCM","RM","LWB","LDM","CDM","RDM","RWB","LB","LCB","CB","RCB","RB"]
for st in stats: df[st] = df[st].replace([np.nan], df[st].mean())
df.head()
```

## Predicción
La predicción del valor del jugador se hará entrenando un modelo de regresión lineal tal y como se vio en clase:

```python
val = df.pop("Value")
```

Se utiliza 1/3 de la muestra para el test

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=31)
```

```python
len(X_train)
```

Entrenamos el modelo:

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

Finalmente obtenemos una métrica 𝑅2 para la regresión

```python
preds = reg.predict(X_test)
r2_score(preds, y_test)
```

Segun la métrica obtenemos un 96.4322% de fiabilidad lo cual es bastante aceptable.
